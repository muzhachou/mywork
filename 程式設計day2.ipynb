````python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
````

````python
π=np.pi
````

````python
θ=np.linspace(0,2*π,500)
````

````python
r=3
x= r*np.cos(θ)
y = r*np.sin(θ)
````

````python
plt.plot(x,y)
````

````python
ax=plt.gca()
ax.set_aspect('equal')
plt.plot(x,y)
````

````python
r=1- np.cos(θ)
x=r*np.sin(θ)
y=r*np.cos(θ)
````

````python
ax=plt.gca()
ax.set_aspect('equal')
plt.plot(x,y,'y')
````

## while迴圈
while條件式:
    條件符合就執行
    （要小心使用，因為常常忘記設停止條件）

````python
n=1
while n<10:
    print("*"*n)
    n=n+1
````
