```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
θ=np.linspace(-2*np.pi,2*np.pi,500)
```

```python
x=np.sin(θ)-1/2*np.cos(np.abs(3*θ))
```

```python
y=np.cos(θ)-1/2*np.sin(np.abs(3*θ))
```

```python
plt.plot(x,y)
```

```python
l=1-np.cos(θ)
h=l*np.sin(θ)
w=l*np.cos(θ)
k=0.5*h
m=0.5*w-0.5
```

```python
ax=plt.gca()
ax.set_aspect('equal')
plt.plot(h,w,"r")
```
